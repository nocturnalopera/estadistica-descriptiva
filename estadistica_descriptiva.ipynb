```python
def promedio(numeros):  # toma una lista de números como parámetro
    if not numeros:
        return 0.0   # si la lista está vacía, retorna 0 para evitar problemas de división por 0
    suma = 0
    for numero in numeros:
        suma += numero
    return suma / len(numeros) # divide la suma de los elementos por la longitud de los elementos para obtener el promedio
```

```python
def mediana(numeros):
    if not numeros:
        return 0.0 # si la lista está vacía, retorna 0
    lista_ordenada = sorted(numeros) # ordena la lista
    n = len(lista_ordenada)
    medio = n // 2
    if n % 2 == 1:
        return lista_ordenada[medio]  # con unnúmero impar de elementos, la mediana es el elemento del medio
    else:
        return (lista_ordenada[medio - 1] + lista_ordenada[medio]) / 2 # con un número par de elementos;
                                                                      #la mediana es el promedio de los dos elementos del medio
```

```python
def moda(lst):
    if not lst:
        return None  # Si la lista está vacía, no hay moda    
    conteo = {} # usa un diccionario para contar la frecuencia de cada elemento en la lista
    for elemento in lst:
        if elemento in conteo: # recorre el diccionario para encontrar el elemento con la mayor frecuencia
            conteo[elemento] += 1 
        else:
            conteo[elemento] = 1    
    max_conteo = 0
    valor_moda = None
    for clave, valor in conteo.items():
        if valor > max_conteo:
            max_conteo = valor
            valor_moda = clave    
    return valor_moda
```

```python
def rango(lst):
    if not lst:
        return None
    return max(lst) - min(lst)
```

```python
def varianza(datos):
    if not datos:
        return 0 
    n = len(datos)
    media = sum(datos) / n
    suma_cuadrados = sum((x - media) ** 2 for x in datos)
    return suma_cuadrados / n
```

```python
def desviacion_estandar(lst):
    if not lst or len(lst) == 1:
        return 0.0  # la desviación estándar de una sola observación es 0
    med = promedio(lst) # calcula la media usando el promedio
    varianza = sum((x - med) ** 2 for x in lst) / len(lst)  # la varianza poblacional
    return varianza ** 0.5
```

```python
def dma(lst):
    if not lst:
        return None
    med = mediana(lst) # calcula la mediana de la lista usando la función de mediana
    desviaciones = [abs(x - med) for x in lst] #calcula las desviaciones absolutas de cada elemento respecto a esa mediana
    return mediana(desviaciones)
```

```python
def iqr(datos):
    datos_ordenados = sorted(datos) # se ordenan los datos
    n = len(datos_ordenados)
    
    def cuartil(pos): # función auxiliar para calcular los cuartiles
        if pos.is_integer(): # si la posición es entera, tomamos directamente el valor
            return datos_ordenados[int(pos) - 1]
        else:
            inferior = datos_ordenados[int(pos) - 1] # si no es entero, se iterpola entre los dos valores más cercanos
            superior = datos_ordenados[int(pos)]
            return (inferior + superior) / 2
    q1_pos = (n + 1) * 0.25 # posiciones de q1 y q3 usando la fórmula (n+1)*p
    q3_pos = (n + 1) * 0.75
    q1 = cuartil(q1_pos)
    q3 = cuartil(q3_pos)
    return q3 - q1
```

```python
def percentil(datos, p):
    if not datos:
        return None
    if p < 0 or p > 100:
        raise ValueError("El percentil debe estar entre 0 y 100")
    datos_ordenados = sorted(datos)
    n = len(datos_ordenados)
    pos = (n - 1) * (p / 100)
    i = int(pos) 
    frac = pos - i 
    if frac == 0:
        return datos_ordenados[i]
    else:
        return datos_ordenados[i] + frac * (datos_ordenados[i + 1] - datos_ordenados[i])
```
